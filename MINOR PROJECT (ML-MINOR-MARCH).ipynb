# Which are the movies with most and least earned revenue?

```python
df2 = df1[['original_title','revenue']]
```

```python
df2
```

```python
df3 = df2.set_index('original_title')
df3
```

```python
dict_data = df3.to_dict()
dict_data
```

```python
new_dict = dict_data['revenue']
```

```python
new_dict
```

```python
sorted_values = sorted(new_dict.values()) # Sort the values
sorted_dict = {}

for i in sorted_values:
    for k in new_dict.keys():
        if new_dict[k] == i:
            sorted_dict[k] = new_dict[k]
            break

```

```python
print(sorted_dict)
```

```python
least = list(sorted_dict)[0]
print("Least earn revenue :",least)
most = list(sorted_dict)[-1]
print("Most earn revenue :",most)
```

# What is the average number of words in movie titles between the year 2000-2005?

```python
df1
```

```python
df1['release_year'].value_counts()
```

```python
df4 = df1[df1['release_year'] <= 2005]
df4['release_year'].value_counts()
```

```python
df4 = df4[df4['release_year'] >= 2000]
df4['release_year'].value_counts()
```

```python
df4[['original_title','release_year']]
```

```python
str_dict = df4['original_title'].to_dict()
str_dict
```

```python
d2 = df4['original_title'].shape[0]
```

```python
list1 = []
for word in str_dict.values():
    list1.append(word)
```

```python
res = 0
for word1 in list1:
    res = res + (len(word1.split()))
```

```python
avg_words = res/d2
```

```python
print("The average number of words in movie title in between year 2000-2005 :",avg_words)
```
